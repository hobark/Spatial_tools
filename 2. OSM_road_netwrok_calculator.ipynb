```python
df = pd.DataFrame(['primary','secondary','tertiary'], columns =["roads"])

for x in range (0,13136):
    shape = data['geometry'][x]
    length = osm_road_length.get(shape)
    primary = length['length'].get('primary',"empty")
    secondary = length['length'].get('secondary',"empty")
    tertiary = length['length'].get('tertiary',"empty")
    append_list = [primary,secondary,tertiary]
    df[x] = append_list
```

```python
print(df.transpose())
```

```python
df2 = df.transpose()
df2.to_csv('UCDB_global_road_osm_110821.csv', encoding='utf-8')
```
