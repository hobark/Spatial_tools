```python
df = pd.DataFrame(['primary','secondary','tertiary'], columns =["roads"])

for x in col_one_list:
    x = int(x-1)
    shape = data2['geometry'][x]
    try:
        length = osm_road_length.get(shape)
        primary = length['length'].get('primary',0)
        secondary = length['length'].get('secondary',0)
        tertiary = length['length'].get('tertiary',0)
        append_list = [primary,secondary,tertiary]
    except ValueError:
        append_list = [0,0,0]
    df[x] = append_list
```
